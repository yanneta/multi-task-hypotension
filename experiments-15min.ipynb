```python
train, valid = get_data_gap(PATH, gap="15min")
```

```python
norm_dict, care2id, id2index, num_subjects = stats_dict(train)
```

```python
train_ds = MultiTask(train, norm_dict, id2index, care2id)
valid_ds = MultiTask(valid, norm_dict, id2index, care2id, train=False)
```

```python
batch_size = 5000
train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
valid_dl = DataLoader(valid_ds, batch_size=batch_size)
```

```python
def multiple_runs(num_subjects, C=1/5, basename="model_15min", n=10):
    prev_val_r2 = 0
    prev_val_r2S = 0
    for i in range(n):
        model = EventModel0(num_subjects).cuda()
        modelS = EventModel0(num_subjects).cuda()
        
        optimizer = torch.optim.Adam(model.parameters(), lr=0.04, weight_decay=1e-5)
        optimizerS = torch.optim.Adam(modelS.parameters(), lr=0.04, weight_decay=1e-5)
        prev_val_r2, prev_val_r2S = train_epochs(model, modelS, optimizer, optimizerS, train_ds, valid_dl,\
                                                 basename, prev_val_r2, prev_val_r2S,\
                                                 max_lr=0.04, epochs = 30, C=C)
        
    return prev_val_r2, prev_val_r2S    
```
